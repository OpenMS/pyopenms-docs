# Query [MSExperiment](https://pyopenms.readthedocs.io/en/latest/apidocs/_autosummary/pyopenms/pyopenms.MSExperiment.html) with MassQL

MassQL is a powerful, SQL-like query language for mass spectrometry
data. For further information visit the [MassQL
documentation](https://mwang87.github.io/MassQueryLanguage_Documentation/).

MS data from a
[MSExperiment](https://pyopenms.readthedocs.io/en/latest/apidocs/_autosummary/pyopenms/pyopenms.MSExperiment.html)
can be exported to
[MS1](https://pyopenms.readthedocs.io/en/latest/user_guide/glossary.html#term-MS1)
and
[MS2](https://pyopenms.readthedocs.io/en/latest/user_guide/glossary.html#term-MS2)
dataframes, which can be queried directly with the `massql` module.

**pyopenms.MSExperiment.get_massql_df()**  
Exports data from
[MSExperiment](https://pyopenms.readthedocs.io/en/latest/apidocs/_autosummary/pyopenms/pyopenms.MSExperiment.html)
to pandas DataFrames to be used with MassQL.

Both dataframes contain the columns: 'i': intensity of a peak 'i_norm':
intensity normalized by the maximun intensity in the spectrum
'i_tic_norm': intensity normalized by the sum of intensities (TIC) in
the spectrum 'mz': mass to charge of a peak 'scan': number of the
spectrum 'rt': retention time of the spectrum 'polarity': ion mode of
the spectrum as integer value (positive: 1, negative: 2)

The
[MS2](https://pyopenms.readthedocs.io/en/latest/user_guide/glossary.html#term-MS2)
dataframe contains additional columns: 'precmz': mass to charge of the
precursor ion 'ms1scan': number of the corresponding
[MS1](https://pyopenms.readthedocs.io/en/latest/user_guide/glossary.html#term-MS1)
spectrum 'charge': charge of the precursor ion

**Returns:**

ms1_df : **pandas.DataFrame**

peak data of
[MS1](https://pyopenms.readthedocs.io/en/latest/user_guide/glossary.html#term-MS1)
spectra

ms2_df : **pandas.DataFrame**

peak data of
[MS2](https://pyopenms.readthedocs.io/en/latest/user_guide/glossary.html#term-MS2)
spectra with precursor information

**Example:**

Load an example file into a
[MSExperiment](https://pyopenms.readthedocs.io/en/latest/apidocs/_autosummary/pyopenms/pyopenms.MSExperiment.html)
and get the
[MS1](https://pyopenms.readthedocs.io/en/latest/user_guide/glossary.html#term-MS1)
and
[MS2](https://pyopenms.readthedocs.io/en/latest/user_guide/glossary.html#term-MS2)
data frames for a MassQL query.

```
import pyopenms as oms
from massql import msql_engine

from urllib.request import urlretrieve

url = "https://raw.githubusercontent.com/OpenMS/pyopenms-docs/master/src/data/"

urlretrieve(url + "small.mzML", "small.mzML")

# load MSExperiment
exp = oms.MSExperiment()
oms.MzMLFile().load("small.mzML", exp)

# get MS1 and MS2 dataframes
ms1_df, ms2_df = exp.get_massql_df()

ms1_df.head()
```

Run a query on `ms1_df` and `ms2_df`. If you don't pass the data frames
`massql_engine.process_query` will read data from the given file name.

```
# Executing Query
results_df = msql_engine.process_query(
    "QUERY scaninfo(MS1DATA) WHERE RTMIN=16",
    "small.mzML",
    ms1_df=ms1_df,
    ms2_df=ms2_df,
)

results_df.head()
```
